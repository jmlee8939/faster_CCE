```python
pen_even = data.iloc[0:7,:]
pen_odd = data.iloc[8:15,:]
iris = data.iloc[16:23,:]
seeds = data.iloc[24:31,:]
usps_even = data.iloc[32:39,:]
usps_odd = data.iloc[40:47,:]
wine = data.iloc[48:,:]
```

```python
## iris

fig, ax1 = plt.subplots(figsize=(10,5))
plt.rc('xtick',labelsize=15)
plt.rc('ytick',labelsize=12)
plt.title('Iris',fontsize=25)
plt.grid(color='black', linestyle='--', linewidth=0.5, axis='y')
ax2 = ax1.twinx()
ax1.set_ylim((0,1))
ax1.set_xlabel('epsilon', fontsize=20)
ax1.set_xticklabels(['','0','1e-6',
                     '5e-6','1e-5',
                     '5e-5','1e-4',
                     '5e-4'])
ax1.set_ylabel('NMI',fontsize=20)
ax2.set_ylim((0,2000))
ax2.set_ylabel('Iteration',fontsize=20)
graph1 = ax1.bar(iris.index, iris.NMI, width = 0.8,color = ['gray'], edgecolor='black')
patterns = ('/', '', '', '', '', '', '')
labels = ('NMI - CCE', 'NMI - faster CCE')
for bar, pattern, label in zip(graph1, patterns, labels):
    bar.set_hatch(pattern)
    bar.set_label(label)
graph2 = ax2.plot(iris.index,iris['iteration'].values, color='red',
         linestyle='-', marker='o', linewidth=2.0, markersize=10, label='Iteration',
                   alpha=0.8)
fig.legend(bbox_to_anchor = (0.9,0.88) , fontsize=12)
plt.savefig('iris.png')
```
