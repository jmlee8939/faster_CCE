```python
def center_number(S):
    n = S.shape[0]
    center_id = list()
    S_diag = np.diag(S)
    S_res = S - np.diag(S_diag)
    for j in range(n):
        if S_diag[j] > np.max(S_res[j, :]):
            center_id.append(j)
    return(len(center_id))
```
